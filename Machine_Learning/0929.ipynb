```python
import numpy as np
import pandas as pd

df = pd.read_csv('./data/student_scores_em.csv')

scores = df['english'][:10]
print(scores.values) # [42 69 56 41 57 48 65 49 65 58]

# 평균을 구해봐요
np.mean(scores.values) # 55.000 scores는 Series
scores.mean() # 55.000

# 이상치가 있는 경우에는 중위값(중앙값)을 대푯값으로 사용
np.median(scores) # 56.500

# 이번에는 대푯값으로 절사평균을 사용
# 절사평균은 numpy가 제공하지 않아요
from scipy import stats

data = [1, 100, 3, 4, 2]
trimmed_mean = stats.trim_mean(data, 0.2)
print(trimmed_mean) # 3
```
