```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
# Raw Data Loading
df = pd.read_csv('./data/students_scores_400.csv')
# print(df.shape) # (400, 2)

# 성적데이터만 가져올꺼예요. ndarray로 가져와요.
scores = df['score'].values

# 전체 데이터에 대한 분포를 살펴보아요!(히스토그램)
plt.hist(scores,
         bins = 100,
         range = (0, 100),
         density = True) # density = True로 설정하면 개수가 아니라 빈도로 표현

# 5.5%의 확률로 80점이라는 표본데이터를 얻을 수 있어요
plt.show()
```

```python
# 무작위 추출로 표본 크기 20인 표본을 추출
# 10000번 추출 (표본을 10000개 생성)
# 표본 평균을 계산하는 작업을 각 표본에 대해서 수행

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
# Raw data Loading
df = pd.read_csv('./data/students_scores_400.csv')

scores = df['score'].values # 성적데이터만 가져와요!

# 리스트로 만들꺼예요 -> List Comphreension을 이용해서 만들어보아요
# 10000개의 각 샘플(샘플크기는 20)의 평균을 구해서 리스트로 생성
sample_means = [np.random.choice(scores, 20).mean() for _ in range(10000)]
# 변수는 있는데 쓰지 않을 때 _ 사용

plt.hist(sample_means,
         bins=100,
         range=(0, 100),
         density=True)

plt.xlim(60,90)
plt.show()

# 결과를 확인해 봤더니.... 표집분포에 대한 분포형태는 정규분포
# 이유는 => 중심극한정리에 의해서 정규분포

# 중심극한정리 ->
# 독립적인 확률변수가 여러개 있을 때, 
# 각 확률변수의 분포화는 상관없이
# 확률변수의 합은 정규분포를 따르게된다.
```
